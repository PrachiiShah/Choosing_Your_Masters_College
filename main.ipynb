```python
#fetchAndSave(url,'data/MS_Colleges.txt')
with open('data/MS_Colleges.txt', 'r', encoding = 'utf-8') as f:
    str = f.read()
soup = BeautifulSoup(str,'html.parser')
```

```python
print(soup.title.string)
```

```python
data = {
    'colleges':[],
    'Location':[],
    'UniversityType':[],
    'No_of_courses_offered':[],
    'TuitionFees_FirstYear':[],
    'Exams_Accepted':[],
    'Work_Experience':[],
    'Scholarships':[],
    'Intake_Session':[]        
       }
```

```python
colleges = soup.find_all(name = 'div', class_='c8ff')
for college in colleges:
   # if not (college.select('div.fd16') or college.select('img') or college.find(name='span',string = ' Shiksha Grade ')):
        print(college.get_text())
        data['colleges'].append(college.get_text())
len(data['colleges'])
```

```python
locations = soup.find_all(name = 'span', class_='_5588')
for location in locations:
    print(location.get_text())
    data['Location'].append(location.get_text())
```

```python
uni_type = soup.find_all(class_='edfa')
for uni in uni_type:
    print(uni.span.find_next_sibling().get_text())
    data['UniversityType'].append(uni.span.find_next_sibling().get_text())
```

```python
course_count = soup.find_all(name='label',string='Courses Offered')
for cc in course_count:
    count = int(cc.find_next_sibling().get_text().removesuffix(' Courses'))
    print(count)
    data['No_of_courses_offered'].append(count)
```

```python
tuition_fees = soup.find_all(name='label',string='1st year Tution Fees')
for fees in tuition_fees:
    fee = fees.find_next_sibling().get_text()
    print(fee)
    data['TuitionFees_FirstYear'].append(fee)
```

```python
exams = soup.find_all(name='label',string='Exams Accepted')
for exam in exams:
    el = []
    exam = exam.find_next_sibling().ul.children
    for e in exam:
        if e.find(class_=['aace ripple dark','aace _555b ripple dark']):
            el.append(e.get_text())
    print(el)
    data['Exams_Accepted'].append(el)
```

```python
scholarship_count = soup.find_all(name = 'label', string = 'No. of Scholarship')
for sc in scholarship_count:
    if sc.find_next_sibling().get_text() == '– / –':
        print(0)
        data['Scholarships'].append(0)
    else:
        res = int(sc.find_next_sibling().get_text())
        print(res)
        data['Scholarships'].append(res)
```

```python
data['Graduation_Marks_Min_Percent'] = []
exp_grad_marks = soup.find_all(name = 'label', string = ['Work Experience', 'Graduation Marks'])
for val in exp_grad_marks:
    exp = val.find_next_sibling().get_text()
    if 'Months' in exp:
        print(exp)
        data['Work_Experience'].append(exp)
        data['Graduation_Marks_Min_Percent'].append(0)
    elif 'Minimum' in exp:
        r= int(exp.removesuffix('%').removeprefix('Minimum'))
        print(r)
        data['Graduation_Marks_Min_Percent'].append(r)
        data['Work_Experience'].append(None)
    else:  
        print('None')
        data['Work_Experience'].append('None')
        data['Graduation_Marks_Min_Percent'].append(0)
```

```python
intake_session = soup.find_all(name = 'label', string = ['Intake Session','Course Duration'])
for session in intake_session:
    sess = session.find_next_sibling().get_text()
    sess_list = []
    if 'Months' not in sess:
        session = session.find_next_sibling().ul.children
        for s in session:
            if not(s.find(string = '+')):
                sess_list.append(s.get_text())
        data['Intake_Session'].append(sess_list)
        print(sess_list)
    elif 'Months' in sess:
        print(sess)
        data['Intake_Session'].append(sess)
```

```python
df = pd.DataFrame(data=data)
```
